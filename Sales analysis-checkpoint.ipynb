```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
 df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

#Exploratory Data analysis

#Gender

```python
df.columns
```

```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gender=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender)

```

#from the above graph we can see that most of the buyers are female and the amount used for purchasing is much higher than male.

#Age

```python
 df.columns
```

```python
sns.countplot(x='Age Group',data=df)
```

```python
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
ay=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ay.containers:
    ay.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

#from the above graph most of the buyers are of the age group between 26-35 are female

#States

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

#most of the orders and amount spent are from Uttar Pradesh, Maharashtra, and Karnataka respectively

#Maritial Status

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

#from the above graph most of the buyers are married wowmen

#Occupation

```python
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_sector=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_sector,x='Occupation',y='Amount')
```

#Product Catogory

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
#from the above graph we can see that most of the products are from Food, Clothing & Apparel and Electronic & Gadgets.
```
